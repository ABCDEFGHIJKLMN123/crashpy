## Files in Python

Handling files is a pre-requisite for a number of Python computing applications. Hence we learn that first.

```python
# Handling Files in Python
file = open("poemRobertFrost.txt", "r", encoding = "utf8")
print("Name of File : ", file.name)
print("File Closed? : ", file.closed)
print("Opening Mode : ", file.mode)
file.close()  # closing a file is crucial
```

```python
# Reading Lines from a File
with open("poemRobertFrost.txt", "r", encoding = "utf8") as file:
    line = file.readline()
    line_count = 1
    while line:
        print("Line {0:2d} \t {1:s}".format(line_count, line.strip()))
        line = file.readline()
        line_count += 1
```

```python
# Reading Lines from a File
with open("poemRobertFrost.txt", "r", encoding = "utf8") as file:
    for line_count, line in enumerate(file):
        print("Line {0:2d} \t {1:s}".format(line_count, line.strip()))
```

```python
# Reading Words from a File
with open("poemRobertFrost.txt", "r", encoding = "utf8") as file:
    for line_count, line in enumerate(file):
        print("Line {0:2d} \t".format(line_count), end = "")
        for word in line.split(" "):
            print(word.strip(), end = "\t")
        print()
```

## Numeric Computation

`NumPy` is one of the most versatile modules in Python. It helps in Numeric and Scientific computations. This library does not come with the standard Python installation. If you want to use `NumPy`, install the package using your Python package manager of choice. I prefer Conda : `conda install numpy`.

```python
# Import NumPy
import numpy as np
```

```python
# Create an Array
x = np.array([1, 2, 3, 4, 5, 6])
print(x, "is of type", x.dtype, "and of size", x.size, "with shape", x.shape)
```

```python
# Reshape Matrix
x = np.array([1, 2, 3, 4, 5, 6])
x = x.reshape(2,3)
print(x, "is of type", x.dtype, "and of size", x.size, "with shape", x.shape)
```

```python
# Array Arithmetic
x = np.arange(1,7).reshape(2,3)
y = np.random.random((2,3))

print(x)
print()
print(y)
print()
print(x + y)
print()
print(x - y)
```

```python
# Array Product
x = np.arange(1,7).reshape(2,3)
y = np.random.random((3,2))

print(x)
print()
print(y)
print()
print(x.dot(y))
```

## Symbolic Computation

`SymPy` is the Python library for Symbolic Computation. You may install the `SymPy` module using Conda : `conda install sympy`.

```python
# Import SymPy
import sympy as sp
```

```python
# Create Symbol and Expression
x = sp.symbols('x')
ep = sp.sin(x) * sp.exp(x)
print(ep, "is of type", type(ep))
```

```python
# Symbolic Derivative
df = sp.diff(ep, x)
print("The derivative of", ep, "is", df)
```

```python
# Symbolic Integration
it = sp.integrate(df, x)
print("The derivative of", df, "is", it)
```

```python
# Differential Equation
t = sp.symbols('t')
y = sp.Function('y')
sp.dsolve(sp.Eq(y(t).diff(t, t) - y(t), sp.exp(t)), y(t))
```

```python
# Pretty Print in LaTeX
z = sp.dsolve(sp.Eq(y(t).diff(t, t) - y(t), sp.exp(t)), y(t))
print(sp.latex(z))
```

```python
# EigenValues of a Matrix
m = np.random.random((2,2))
M = sp.Matrix(m)
print(M)
print()
print(M.eigenvals())
print()
print(M.eigenvects())
```

## Data Analysis

Python is currently the language of choice for Data Analysis; the required libraries are `NumPy`, `Pandas`, `MatPlotLib`, `SeaBorn` and `Scikit-Learn`. In this example, we use the **"Pokemon with stats"** dataset from Kaggle, curated by *Alberto Barradas* (source: https://www.kaggle.com/abcsds/pokemon).

```python
# Import Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
# Read the CSV Data
pkmndata = pd.read_csv('pokemonData.csv')
pkmndata.head()
```

```python
# Check the Data Type
print("Data type : ", type(pkmndata))
print("Data dims : ", pkmndata.shape)
print()
print(pkmndata.dtypes)
```

```python
# Extract only the numeric data variables
numDF = pd.DataFrame(pkmndata[["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]])

# Summary Statistics for all Variables
numDF.describe()
```

```python
# Draw the distributions of all variables
f, axes = plt.subplots(6, 3, figsize=(18, 24))
color_list = ["b", "g", "r", "c", "m", "y"]

count = 0
for var in numDF:
    sb.boxplot(numDF[var], ax = axes[count,0], color = color_list[count])
    sb.distplot(numDF[var], ax = axes[count,1], color = color_list[count])
    sb.violinplot(numDF[var], ax = axes[count,2], color = color_list[count])
    count += 1
```

```python
# Heatmap of the Correlation Matrix
f, axes = plt.subplots(1, 1, figsize=(12, 8))
sb.heatmap(numDF.corr(), vmin = -1, vmax = 1, annot = True, fmt = ".2f")
```

```python
# Draw pairs of variables against one another
sb.pairplot(data = numDF)
```
